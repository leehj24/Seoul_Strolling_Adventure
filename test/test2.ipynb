```python
import pandas as pd

def pick_two(df1: pd.DataFrame, df2: pd.DataFrame) -> pd.DataFrame:
    key_cols = ['자치구', '행정동']

    # 두 컬럼 모두 NaN 인 행은 의미가 없으므로 미리 제거
    df1_clean = df1[key_cols].dropna(how='all').drop_duplicates()
    df2_clean = df2[key_cols].dropna(how='all').drop_duplicates()

    # -----------------------------------------------------------
    # ① 둘 중 하나가 완전히 비어 있을 때  -----------------------
    # -----------------------------------------------------------
    if df1_clean.empty and df2_clean.empty:
        # 두 DF 모두 내용이 없으면 빈 DF 반환
        return pd.DataFrame(columns=key_cols)

    if df1_clean.empty:          # df1 쪽이 전부 NaN
        return df2_clean.head(2).reset_index(drop=True)

    if df2_clean.empty:          # df2 쪽이 전부 NaN
        return df1_clean.head(2).reset_index(drop=True)

    # -----------------------------------------------------------
    # ② 둘 다 내용이 있을 때  ------------------------------------
    # -----------------------------------------------------------
    # 1) 교집합
    common = (
        pd.merge(df1_clean, df2_clean, on=key_cols)
          .drop_duplicates()
    )

    # (A) 교집합이 정확히 2개
    if len(common) == 2:
        return common.reset_index(drop=True)

    # (B) 교집합이 2개보다 많음
    if len(common) > 2:
        top1_df1 = df1_clean.iloc[0]
        top1_df2 = df2_clean.iloc[0]

        # 두 DF의 1위가 동일하면 df2 상위 2개
        if (top1_df1 == top1_df2).all():
            return df2_clean.head(2).reset_index(drop=True)
        # 다르면 각 DF의 1위씩
        return pd.DataFrame([top1_df1, top1_df2])[key_cols].reset_index(drop=True)

    # (C) 교집합이 0‒1개
    candidates = pd.concat([
        common,
        df1_clean.head(1),
        df2_clean.head(1)
    ]).drop_duplicates()

    return candidates.head(2).reset_index(drop=True)


# 사용 예
result = pick_two(df_top10, df2_top10)
result
```

```python
import pandas as pd
import numpy as np

# 1) Haversine 거리 계산 함수
def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return 6371 * 2 * np.arcsin(np.sqrt(a))

# 2) 추천 대상 행정동들의 중심 좌표 준비
regions = result.merge(
    df_top10[['자치구','행정동','위도','경도']],
    on=['자치구','행정동'], how='left'
).rename(columns={'위도':'reg_lat','경도':'reg_lon'})

# 3) 리뷰 데이터에 음식점 좌표(place_name, lat, lon) 합치기
region_sub = df_region[['addr1','mapy','mapx','title']].rename(
    columns={'mapy':'lat','mapx':'lon','title':'place_name'}
)
df_rev_geo = pd.merge(df_review, region_sub, on='addr1', how='inner')

# 4) 행정동별 정렬된 place_name 목록 생성
region_sorted = {}
for _, row in regions.iterrows():
    dong, lat0, lon0 = row['행정동'], row['reg_lat'], row['reg_lon']
    df_dong = df_rev_geo[
        (df_rev_geo['행정동']==dong)&(df_rev_geo['cat1']=='음식')
    ].copy()
    if df_dong.empty:
        region_sorted[dong] = []
        continue
    df_dong['distance_km'] = haversine(
        lon0, lat0,
        df_dong['lon'].values,
        df_dong['lat'].values
    )
    df_sorted = df_dong.sort_values(
        ['긍정점수','distance_km'],
        ascending=[False, True]
    )
    region_sorted[dong] = df_sorted['place_name'].tolist()

# 5) 중복 없이, 각 동에서 5개씩 뽑고 부족하면 다른 동 목록에서 채워서 총 10개 확보
required_per_region = 5
total_required = required_per_region * len(region_sorted)

unique_places = []
seen = set()

# 1차: 동별로 최대 5개씩
for dong, names in region_sorted.items():
    count = 0
    for name in names:
        if name not in seen:
            unique_places.append(name)
            seen.add(name)
            count += 1
        if count == required_per_region:
            break

# 2차: 만약 10개 미만이면, 모든 동의 리스트를 순회하며 남은 슬롯 채우기
if len(unique_places) < total_required:
    for names in region_sorted.values():
        for name in names:
            if name not in seen:
                unique_places.append(name)
                seen.add(name)
            if len(unique_places) == total_required:
                break
        if len(unique_places) == total_required:
            break

# 6) 결과 확인
print(f"▶ 총 추천 음식점 개수: {len(unique_places)}")
print(unique_places)
```

```python
import pandas as pd
import numpy as np

# 1) 벡터화된 Haversine 거리 계산 함수
def haversine(lon1, lat1, lons, lats):
    """
    lon1, lat1: 스칼라 기준점 (float)
    lons, lats: 판다스 Series 또는 numpy array (길이 n)
    반환: 두 점 간 거리 배열 (km)
    """
    # 라디안 변환
    lon1_rad, lat1_rad = np.radians(lon1), np.radians(lat1)
    lons_rad = np.radians(lons)
    lats_rad = np.radians(lats)
    # 차이
    dlon = lons_rad - lon1_rad
    dlat = lats_rad - lat1_rad
    # Haversine 공식
    a = np.sin(dlat/2)**2 + np.cos(lat1_rad) * np.cos(lats_rad) * np.sin(dlon/2)**2
    return 6371 * 2 * np.arcsin(np.sqrt(a))  # km

# 2) df_region에서 음식점 좌표(place_name, lat, lon) 준비
region_sub = df_region[['addr1','mapy','mapx','title']].rename(
    columns={'mapy':'lat','mapx':'lon','title':'place_name'}
)

# 3) df_review + region_sub 병합하여 좌표 포함 리뷰 데이터 생성
df_rev_geo = pd.merge(df_review, region_sub, on='addr1', how='inner')

# 5) 전역 중복 방지 세트
used2 = set()
used3 = set()

rows = []
for init in unique_places:
    base = df_rev_geo[df_rev_geo['place_name'] == init]
    if base.empty:
        continue
    lat0, lon0 = base.iloc[0][['lat','lon']]
    
    # 추천장소2 후보 및 거리 계산 (벡터 방식)
    df2 = df_rev_geo[df_rev_geo['cat1']=='음식'].copy()
    df2['dist1'] = haversine(lon0, lat0, df2['lon'], df2['lat'])
    df2_sorted = df2.sort_values(['긍정점수','dist1'], ascending=[False, True])
    
    # rec2 선택 (5개, 중복.Global 방지)
    rec2, dist1_vals = [], []
    for _, r in df2_sorted.iterrows():
        name, d1 = r['place_name'], r['dist1']
        if name != init and name not in used2:
            rec2.append(name)
            used2.add(name)
            dist1_vals.append(d1)
        if len(rec2) == 5:
            break
    avg_d1 = np.mean(dist1_vals) if dist1_vals else np.nan
    
    # 추천장소3 후보 생성
    cand3 = []
    for p1 in rec2:
        t = df_rev_geo[df_rev_geo['place_name'] == p1].iloc[0]
        lat1, lon1 = t['lat'], t['lon']
        df3 = df_rev_geo[df_rev_geo['cat1']=='음식'].copy()
        df3['dist2'] = haversine(lon1, lat1, df3['lon'], df3['lat'])
        df3 = df3[df3['dist2'] <= 2]
        cand3.append(df3)
    
    if cand3:
        df3_all = pd.concat(cand3)
        df3_sorted = df3_all.sort_values(['긍정점수','dist2'], ascending=[False, True])
        # rec3 선택 (5개, 중복.Global 방지 + init/rec2 제외)
        rec3, dist2_vals = [], []
        for _, r in df3_sorted.iterrows():
            name, d2 = r['place_name'], r['dist2']
            if name not in used3 and name not in rec2 and name != init:
                rec3.append(name)
                used3.add(name)
                dist2_vals.append(d2)
            if len(rec3) == 5:
                break
        avg_d2 = np.mean(dist2_vals) if dist2_vals else np.nan
    else:
        rec3, avg_d2 = [], np.nan
    
    rows.append({
        '추천장소1': init,
        '도보이동km_1': round(avg_d1, 2),
        '추천장소2': rec2,
        '도보이동km_2': round(avg_d2, 2) if not np.isnan(avg_d2) else None,
        '추천장소3': rec3
    })

# 6) 결과 DataFrame
df_routes = pd.DataFrame(rows)
# 7) 리스트 컬럼을 문자열로 변환해 대괄호 제거
df_routes['추천장소2'] = df_routes['추천장소2'].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')
df_routes['추천장소3'] = df_routes['추천장소3'].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')

df_routes
```
