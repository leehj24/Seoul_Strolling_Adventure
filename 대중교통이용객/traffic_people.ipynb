```python
import pandas as pd
import glob

# 파일 경로 지정
path = r'C:\Users\hyunj\machin_prj\대중교통 이용객\버스'
file_list = glob.glob(f"{path}/*.csv")  # 폴더 내의 모든 .xlsx 파일 가져오기

# 모든 파일을 읽어 DataFrame에 추가
dataframes = []
for file in file_list:
    df = pd.read_csv(file,encoding='cp949')  # 엑셀 파일 읽기
    dataframes.append(df)

# 하나의 DataFrame으로 결합
combined_df = pd.concat(dataframes, ignore_index=True)

```

```python
# 1. 숫자형 열만 선택
numeric_cols = combined_df.select_dtypes(include='number').columns

# 2. '역명'별로 숫자형 열 합계 계산
result_bus = combined_df.groupby('역명')[numeric_cols].sum().reset_index()

# 3. 숫자형 열 전체의 합계를 계산해서 '총합' 열 생성
result_bus['총합'] = result_bus[numeric_cols].sum(axis=1)

result_bus = result_bus[['역명', '총합']]
result_bus.head(4)
```
