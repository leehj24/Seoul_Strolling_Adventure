```python
import pandas as pd

file = r'E:\machin-prj\지역\ZB)TourAPI_area_based_seoul_incheon_FINAL.xlsx'
# 엑셀 파일 읽기
df_a = pd.read_excel(file)  # 엑셀 A
df_b = pd.read_excel("카테고리.xlsx")  # 엑셀 B

# 엑셀 B의 '카테고리'와 '코드' 열을 이용하여 매핑 딕셔너리 생성
mapping = dict(zip(df_b['cat1'], df_b['cat1_name']))

# 매핑 정보를 기반으로 '업소명' 열의 내용 중 어느 키워드가 포함되어 있는지 확인 후 코드 반환하는 함수
def get_code(text):
    for category, code in mapping.items():
        if category in str(text):  # 문자열인지 확인
            return code
    return None

# '업소명' 열을 기준으로 새 열 '코드' 추가
df_a['cat1'] = df_a['cat1'].apply(get_code)

# 결과 엑셀 파일로 저장
df_a.to_excel("excel_a_updated.xlsx", index=False)
print("업데이트된 파일이 'excel_a_updated.xlsx'로 저장되었습니다.")
```
